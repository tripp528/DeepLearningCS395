```python
import matplotlib.pyplot as plt
plt.style.use("dark_background")
import pandas as pd
```

### 1. From Scratch

```python
fromscratch_hist= pd.read_csv("output/results/fromscratch_hist.csv")

plt.figure(figsize=(10,10))
plt.plot(fromscratch_hist.epoch, fromscratch_hist.accuracy, label="train accuracy")
plt.plot(fromscratch_hist.epoch, fromscratch_hist.val_accuracy, label="val accuracy")
plt.xlabel("epochs")
plt.legend()
```

### 2. Feature extraction

```python
extract_hist= pd.read_csv("output/results/extract_hist.csv")

plt.figure(figsize=(10,10))
plt.plot(extract_hist.epoch, extract_hist.acc, label="train accuracy")
plt.plot(extract_hist.epoch, extract_hist.val_acc, label="val accuracy")
plt.xlabel("epochs")
plt.legend()
```

### 3. Transfer learning

```python
transfer_hist= pd.read_csv("output/results/transfer_hist.csv")
plt.figure(figsize=(10,10))


plt.plot(transfer_hist.epoch, transfer_hist.accuracy, label="train accuracy")
plt.plot(transfer_hist.epoch, transfer_hist.val_accuracy, label="val accuracy")
plt.xlabel("epochs")
plt.legend()
```
